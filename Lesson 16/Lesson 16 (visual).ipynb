#### Основні команди Matplotlib

```python
plt.plot(x, y, 'b')

plt.xlabel('X')
plt.ylabel('Y')

plt.title('First')
```

#### Декілька кординатних площин на одному полотні

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'g-*') 

plt.subplot(1,2,2)
plt.plot(y, x, 'r-')
```

#### Обєктно-орієнтовний підхід

Об'єкт plt.subplots () працює як автоматичний менеджер осей.

```python
fig, axes = plt.subplots(nrows=2, ncols=3)
```

```python
fig
```

```python
axes
```

```python
axes[0][1]
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

plt.tight_layout()
```

```python
help(plt.tight_layout)
```

```python
fig, ax = plt.subplots(2, 2)


ax[0][0].plot(x, x**2)


ax[1][1].plot(x, x**(1/2))

ax[0][1].plot(x, y, 'r')

plt.tight_layout()
```

```python
ax
```

#### Легенда, заголовок, мітки даних

```python
fig, ax = plt.subplots()

ax.set_title("Заголовок");
ax.set_xlabel("x")
ax.set_ylabel("y")

ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.plot(x, x**4, label="x**4")

ax.legend(loc=2) #параметр loc визначає розміщення легенди
```

#### Розмір полотна

```python

fig, axes = plt.subplots(figsize=(4,4))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Заголовок');

# fig = plt.figure(figsize=(8,1), dpi=100)
```

#### Відображення ліній осей

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))


axes[0].plot(x, x**2, x, x**3, lw=1)
axes[0].grid(True)


axes[1].scatter(x,y)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```

### KDE plot

Kernel Density Estimate
В статистиці, я́дрова оці́нка густини́ розпо́ділу  — це непараметричний метод оцінки функції густини випадкової величини за вибіркою. Ядрова оцінка густини є важливою задачею згладжування даних; при застосуванні методу судження щодо статистичних властивостей популяції здійснюється на базі скінченної вибірки.

```python
sns.kdeplot(salaries[salaries['Year']==2013]['TotalPay'], color='b', shade=True, label='2013') 
sns.kdeplot(salaries[salaries['Year']==2012]['TotalPay'], color='r', shade=True, label='2012') 
```

```python
salaries['JobTitle'] = salaries.JobTitle.str.lower() 

top_ten_occupations = salaries.JobTitle.value_counts().sort_values(ascending=False).head(10).index
top_ten_occupations
```

```python
salaries_averages_by_occupation = (salaries[salaries.JobTitle.isin(top_ten_occupations)]
                                   .groupby('JobTitle')[['BasePay', 'Benefits', 'OvertimePay']]
                                   .aggregate('mean')
)

ax = salaries_averages_by_occupation.plot(kind='barh', figsize=(8,8))

ax.set_xlabel('Mean Pay')
```

```python
salary_percents = salaries_averages_by_occupation.div(salaries_averages_by_occupation.sum(1), 
                                                      axis=0)
  
ax = salary_percents.plot(kind='bar', stacked=True)
```
