```python
from pprint import pprint
DB = {}

for row in open('new_file.txt'):
    code = row.split('|')[0]
    name = row.split('|')[1]
    isAviable = row.split('|')[2]
    price = int(row.split('|')[3].rstrip(' грн\n'))

    DB[code] = {
        'name': name,
        'isAviable': isAviable,
        'price': price
    }
pprint(DB)
    
```

```python
len(DB)
```

```python
for i in DB:
    print(i, DB[i])
```

```python
qty = 0
for k, v in DB.items():
    if v['isAviable'] == 'В наличии':
        qty += 1

qty
```
